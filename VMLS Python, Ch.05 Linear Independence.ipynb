# 5.1,2 Linear Independence, Basis

Linear independence is a fundamental concept in linear algebra because it ensures that each piece of information in a set of data is unique and not redundant. Its primary importance lies in **ensuring the uniqueness of representations** and the **solvability of mathematical models**.

### **1. Avoiding Ambiguity through Unique Representation**
You are correct that linear independence is used to avoid ambiguity. If a collection of vectors is linearly independent, any vector that can be built from them can be expressed in **exactly one unique way**.
*   **Deducing Coefficients:** If vectors are linearly independent, we can uniquely infer or deduce the specific weights (coefficients) used to create any linear combination of them.
*   **Basis of a Space:** A **basis** is a collection of linearly independent vectors that act as the minimum set of "building blocks" for a space. Because they are independent, every point in that space has a unique "address" or coordinate relative to that basis.

### **2. Eliminating Redundancy**
A set of vectors is **linearly dependent** if at least one vector in the set can be expressed as a combination of the others. In data science, this means that the specific feature or variable is "redundant" because its information is already captured by other features. Linear independence ensures that every vector provides a new, unique direction or dimension to the dataset.

### **3. Mathematical Solvability and Inversion**
Linear independence is the mathematical requirement for many operations to "work" without errors:
*   **Matrix Invertibility:** A square matrix is invertible (meaning we can "undo" its operation) if and only if its columns are linearly independent.
*   **Solving Equations:** A matrix has a **left inverse** (allowing us to solve over-determined systems of equations) if and only if its columns are linearly independent.
*   **Least Squares:** In data fitting and regression, the standard algorithm to find the "best fit" requires the columns of the data matrix to be linearly independent to ensure a unique, well-defined solution exists.

### **4. Defining Dimensionality**
Linear independence defines the "size" of a space. The **independence-dimension inequality** states that in an $n$-dimensional space, you can have at most $n$ linearly independent vectors. This helps scientists understand the true dimensionality of their data—if you have 100 features but only 10 are linearly independent, your data actually lives in a much smaller 10-dimensional subspace.

# 5.4 Gram-Schmidt

```python
def gram_schmidt(a, tol=1e-10):
    q = []
    for i in range(len(a)):
        qtilde = a[i]
        for j in range(i):
            qtilde = qtilde - np.inner(q[j],a[i])*q[j]
        if npl.norm(qtilde) < tol:
            print("Vectors linearly dependent")
        q.append(qtilde/npl.norm(qtilde))
    return q


a = np.array([[-1,1,-1,1],[-1,3,-1,3],[1,3,5,7]])
```

```python
q = gram_schmidt(a)
q, npl.norm(q[0]), np.inner(q[0],q[1]),np.inner(q[0],q[2]),npl.norm(q[2]), np.inner(q[1],q[2]), npl.norm(q[2])
```

## 5.5 Linear Independence and Data Science

While checking for **linear independence** is a critical task in data science, the sources suggest that your "first job" or initial steps typically involve broader preparation, like **feature representation** and **standardization**. However, understanding the linear independence of your dataset is vital before you begin building certain models.

Here is how you should interpret the role of linear independence in your workflow:

### **1. Why you check for it: Identifying Redundancy**
Checking for linear independence is essentially a check for **redundancy**. If a feature (a column in your data matrix) is a linear combination of others, it provides no unique information and is considered redundant.
*   **Unique Representation**: If your vectors are linearly independent, any data point can be expressed in exactly one unique way as a combination of them.
*   **Model Solvability**: Many essential data science tools, such as **least squares regression**, mathematically require the columns of your data matrix to be linearly independent to ensure a unique, well-defined solution exists.

### **2. How you check for it: The Gram–Schmidt Algorithm**
The standard mathematical tool to determine if a list of vectors is linearly independent is the **Gram–Schmidt algorithm**.
*   **The Process**: It systematically transforms your vectors into an orthonormal set.
*   **The Test**: If the algorithm "quits early" (meaning it produces a zero vector during its iterations), it has detected that your vectors are linearly dependent.

### **3. The Actual "First Step": Standardization**
According to the sources, the **first step in feature engineering** is often **standardizing** (or $z$-scoring) your features. This ensures that all attributes have a mean near zero and a standard deviation around one, which prevents features with large numerical ranges from disproportionately influencing the model.

### **4. When you don't have to check first: Regularization**
In modern data science, you don't always have to manually strip out dependent features before you start. **Regularized least squares** (such as ridge regression) is a powerful tool that allows you to fit a model even if you have more features than data points (a wide matrix).
*   A major "side benefit" of regularization is that the columns of the resulting mathematical structure are **always linearly independent**, regardless of whether your original data matrix was dependent or not.

In summary, while you must ensure linear independence for your models to be stable and unique, your initial focus is usually on **representing your data correctly as vectors** and **standardizing them** for analysis.
