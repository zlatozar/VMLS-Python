## Linear Algebra

The essence of linear algebra: `Linear functions ↔ Inner products ↔ Weight vectors`

The ability to determine a linear function from basis evaluations is the inner product representation in action. The formula `f(x) = Σ f(eᵢ)·xᵢ` is how superposition and linearity let us predict any output from just n experiments.

This is why linear models are so powerful in data science: once you learn the weights (through n "experiments" on your training data), you can predict for any new input with just a dot product.

So how do we find the weight vector w? This is where the distinction between mathematical theory and statistical practice becomes critical. This is where the distinction between mathematical theory and statistical practice becomes critical.

## How We Actually Find w in Data Science

1. Collect Observed Data

We gather m examples `(x⁽¹⁾, y⁽¹⁾), (x⁽²⁾, y⁽²⁾), ..., (x⁽ᵐ⁾, y⁽ᵐ⁾)` where:

`x⁽ⁱ⁾` is a feature vector (e.g., age, income, clicks)
`y⁽ⁱ⁾` is the observed output (e.g., purchase amount, click probability)

2. Assume a Linear Model

We assume `y ≈ w·x` (or with bias: `y ≈ w·x + b`)

3. Solve for w Using the Data

We have m equations:

```
w₁x₁⁽¹⁾ + w₂x₂⁽¹⁾ + ... + wₙxₙ⁽¹⁾ ≈ y⁽¹⁾
w₁x₁⁽²⁾ + w₂x₂⁽²⁾ + ... + wₙxₙ⁽²⁾ ≈ y⁽²⁾
...
w₁x₁⁽ᵐ⁾ + w₂x₂⁽ᵐ⁾ + ... + wₙxₙ⁽ᵐ⁾ ≈ y⁽ᵐ⁾
```
In matrix form: `Xw ≈ y`

We solve for w using methods like:

- Least squares: `w = (XᵀX)⁻¹Xᵀy` (if `XᵀX` is invertible)
- Gradient descent: Iteratively minimize `||Xw - y||²`

We don't get to choose the basis vectors in most data science problems. Instead, we work with whatever features nature (or our data collection process) gives us, and we estimate the linear function from observed examples.

The mathematical theory tells us that if we could experiment with basis vectors, we could determine w exactly with n measurements. In practice, we do the next best thing: collect many observations at arbitrary points and solve for w statistically.

This is why linear regression is so fundamental — it's how we approximate the ideal "basis experiment" approach when we can only observe the world as it naturally presents itself.

## What is a Unit Vector?

A unit vector is a vector that has a magnitude (or length) of exactly 1.

It doesn't matter what direction it points—north, east, or somewhere in between—if its length is 1, it's a unit vector.

Think of it like a standard direction arrow. No matter which way it points, the arrow itself is exactly one unit long (where that "unit" is the scale of your coordinate system).

## How to Get a Unit Vector (Normalization)

Any non-zero vector can be converted into a unit vector. This process is called normalization. You simply divide the vector by its own magnitude.

If v is a vector and `||v||` is its magnitude, then its unit vector û (often pronounced "v-hat") is:

`û = v / ||v||`

This new vector û points in the exact same direction as v, but its length is now 1.

### Once vector

A vector where every component is 1. In linear models like Linear Regression, Logistic Regression, or Neural Networks, you often add a "bias" term (the intercept).
To handle this elegantly with matrix math, you prepend a column of ones to your feature matrix X. When you multiply this column by the bias weight w₀, it simply adds w₀ to every prediction.

`y = X·w + b` becomes `y = [𝟙 , X] · [b, w]ᵀ`

This turns the bias term into just another learnable weight, making the math uniform and efficient.

#### Linear Combination

````python
a, b = np.array([1, 2]), np.array([3, 4])
alpha, beta = -.5, 1.5
a, b, alpha, beta
````

````python
c = alpha*a + beta*b
c
````

````python
coef = ([-.5, 1.5])
vecs = [np.array([1, 2]), np.array([3, 4])]


def lincomb(coeff, vectors):
    n = len(vectors[0])
    a = np.zeros(n)
    for i in range(len(vectors)):
        a += coeff[i] * vectors[i]
    return a


# or, you can use a List Comprehension:

def lincomb_LC(coeff, vectors):
    return [coef[i] * vecs[i] + coef[i + 1] * vecs[i + 1]
            for i in range(len(coef) - 1)]
    # or: sum([coef[i] * vecs[i] for i in range(len(vecs))])


lincomb(coef, vecs), lincomb_LC(coef, vecs)
````
