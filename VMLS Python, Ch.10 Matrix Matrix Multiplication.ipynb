# 10.1 Matrix-Matrix Multiplication

The inner product (also called the dot product) is the fundamental mathematical operation for measuring similarity, calculating scores, and aggregating information.

Measure of Similarity and Alignment The inner product reveals the geometric relationship between two vectors.
1. Orthogonality: If the inner product is zero, the vectors are perpendicular (orthogonal), meaning they have no "shared" components.
2. Alignment: If the inner product equals the product of the vectors' magnitudes (a T
 b=∥a∥∥b∥), the vectors are perfectly aligned.
Correlation: In statistics, the correlation coefficient (ρ)—which tells us how two variables vary together—is essentially the inner product of standardized, de-meaned data vectors

```python
A, B = np.array([[-1.5, 3, 2], [1, -1, 0]]), np.array([[-1, -1], [0, -2], [1, 0]])
C = np.matmul(A, B)  #J: A*B
C
```
