# 6.1 Matrices

Picking out specific columns and rows from a matrix is a fundamental operation in linear algebra and data science, often achieved through the use of **standard unit vectors** and **selector matrices**.

### **1. Picking a Single Column or Row**
The most efficient way to isolate a single column or row mathematically is by multiplying the matrix by a standard unit vector ($e_i$).
*   **To pick the $j$-th column**: Multiply the matrix $A$ by the $j$-th unit vector $e_j$. The identity $A e_j = a_j$ extracts the $j$-th column of the matrix.
*   **To pick the $i$-th row**: Multiply the transpose of the matrix ($A^T$) by the $i$-th unit vector $e_i$. The result, $A^T e_i$, is an $n$-vector that represents the **$i$-th row of $A$, transposed**.

### **2. Using Selector Matrices**
A **selector matrix** is a matrix where each row is a transposed unit vector. These are used to "pick out" or reorder specific components of a vector or matrix:
*   **Slicing**: An $r:s$ slicing matrix can be used to extract a contiguous "slice" of data, such as rows $p$ through $q$ or columns $r$ through $s$.
*   **Permutation**: A permutation matrix, where each row and column is a unit vector, can be used to reorder the rows or columns of a matrix.
*   **Down-sampling**: In time series data, a selector matrix can pick out every other entry (e.g., $x_1, x_3, x_5 \dots$), effectively "down-sampling" the data.

```python
# running sum
def running_sum(n):
    S = np.zeros((n, n))
    for i in range(n):
        for j in range(i + 1):  # 0 indexing fix to get diags = 1
            S[i, j] = 1

    return S
```

```python
# Vandermonde matrices are:
# "a type of matrix that arises in the polynomial least squares fitting,
# Lagrange interpolating polynomials (Hoffman and Kunze p. 114),
# and the reconstruction of a statistical distribution from the
# distribution's moments [mean, variance, skewness, kurtosis (sharpness)]"
def vander(t, n):
    m = len(t)
    V = np.zeros((m, n))
    for i in range(m):
        for j in range(n):
            V[i, j] = t[i] ** (j)
    return V


# ie:
#  vander = lambda t,n : np.matrix([[i**(j) for i in t] for j in range(n)]).transpose()
# or
#  vander = lambda t,n : np.vstack([np.array(t)**i for i in range(n)]).transpose()
vander([-1, -0, 0.5, 1], 5)
```

### Input output interpretation of matrix vector multiplication

 We can interpret the relation y= Ax, with A as a
m×n matrix, as a mapping from the n-vector x to the m-vector y. In this
context we might think of x as an input, and y as the corresponding output. We can interpret Aij as the factor by which yi depends on xj.

### Finding matrix

If you have a transformation that you know can be represented as y=Ax, but you don't know the entries of A, you can "extract" its columns one by one, by applying the transformation to each unit vector (e1, e2, ... en).

### Data scientists job

In the relationship $y = Ax$, the $n$-vector **$x$ is the input** (also called the argument, regressor, or independent variable) and the $m$-vector **$y$ is the output** (also called the result, outcome, or response).

When data scientists or engineers "search" for a transformation matrix $A$, they are performing a task called **matrix estimation** or **model fitting**. To find $A$, they need a dataset consisting of both the inputs they provided (or observed) and the resulting outputs they collected.

### **1. What is the input?**
In this context, the input $x$ represents the **independent variables** or **features**. Depending on the application, the input can be:
*   **Feature Vectors**: Measurements or attributes pertaining to a single object, such as a patient's age and blood pressure or a house's square footage and number of bedrooms.
*   **Standard Unit Vectors ($e_i$)**: In "black-box" experiments, engineers often use standard unit vectors as inputs (e.g., applying a 1-unit load to only one specific spot on a bridge) to directly "extract" the columns of the matrix $A$.
*   **Control Actions**: In engineering tasks like robotics or climate control, the input $x$ represents the **actions** taken, such as engine thrust or heating power levels.

### **2. How $A$ is found from the data**
Engineers typically collect a set of $N$ examples of input-output pairs: $(x^{(1)}, y^{(1)}), \dots, (x^{(N)}, y^{(N)})$.
*   **The Data Matrix ($X$)**: All the **inputs** are collected into a matrix $X$.
*   **The Output Matrix ($Y$)**: All the **collected outputs** are gathered into a matrix $Y$.

The goal is then to find a matrix $A$ such that $AX \approx Y$. This is typically solved using **matrix least squares**, where the engineer searches for the matrix $A$ that minimizes the total squared error (residual) between the predicted outputs ($Ax$) and the actual collected data ($y$).

### **3. Interpretation of the found matrix $A$**
Once the matrix $A$ is found, each entry $A_{ij}$ has a specific meaning: it represents the **sensitivity** of the $i$-th output component to the $j$-th input component. This allows the engineer to understand exactly how a change in one "input" feature will impact every "output" result.
